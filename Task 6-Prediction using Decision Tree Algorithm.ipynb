### Prediction using Decision Tree Algorithm
#### Create the Decision Tree classifier and visualize it graphically.

### Importing Libraries

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Loading the data

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
df.Species.unique()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

### Data Visualization

```python
plt.bar(df['Species'],df['SepalLengthCm'])
plt.bar(df['Species'],df['SepalWidthCm'],color='green')
plt.legend()
```

```python
plt.bar(df['Species'],df['PetalLengthCm'])
plt.bar(df['Species'],df['PetalWidthCm'],color='green')
plt.legend()
```

```python
sns.pairplot(df,hue='Species')
```

```python
X=df.iloc[:,:-1].values
X
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
y=le.fit_transform(df['Species'])
y
```

### Spliting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.1)
```

### Predicting results

```python
from sklearn import tree
model=tree.DecisionTreeClassifier(criterion = 'gini')
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

##### Using confusion matrix we can identify which data was predicted wrongly

### Visualizing the tree

```python
fig = plt.figure(figsize=(16,8))
f=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
c=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
tree.plot_tree(model, filled = True,class_names=c,feature_names=f);
plt.show()
```
